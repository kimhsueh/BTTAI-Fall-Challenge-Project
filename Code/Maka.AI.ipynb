```python
# Assuming you have a DataFrame 'df' with both numeric and non-numeric columns

# Specify the name of the specific column you want to correlate with
target_column = 'Prepayment Penalty Indicator'

# Select only the numeric columns
numeric_columns = df.select_dtypes(include=['number'])

# Check for non-numeric values and convert columns to numeric
numeric_columns = numeric_columns.apply(pd.to_numeric, errors='coerce')

# Calculate the correlations between the target column and numeric columns
correlations = numeric_columns.corrwith(df[target_column])

# Sort the correlations in descending order and get the top five
top_correlations = correlations.abs().sort_values(ascending=False).head(5)

# Print the top five correlations
print("Top five correlations with '{}' column:".format(target_column))
print(top_correlations)






```
